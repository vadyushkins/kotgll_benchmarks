```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="ticks", palette="pastel")
sns.set(rc={'figure.figsize':(12,10)})
```

```python
graphs = [
    "skos",
    "generations",
    "travel",
    "univ",
    "atom",
    "biomedical",
    "foaf",
    "people",
    "funding",
    "wine",
    "pizza",
    "core",
    "pathways",
    "enzyme",
    "eclass",
    "go_hierarchy",
    "go",
    "geospecies",
    "taxonomy",
]
```

```python
df_cfg_g2_without_sppf = pd.concat(
    [
        pd.DataFrame({
            "graph": graph,
            "time_s": pd.read_csv(f"result/{graph}_cfg_g2_without_sppf.csv", header=None).divide(1e9)[0].to_numpy(),
        })
        for graph in graphs
    ],
)
df_rsm_g2_without_sppf = pd.concat(
    [
        pd.DataFrame({
            "graph": graph,
            "time_s": pd.read_csv(f"result/{graph}_rsm_g2_without_sppf.csv", header=None).divide(1e9)[0].to_numpy(),
        })
        for graph in graphs
    ],
)
dff = pd.concat(
    {
        "cfg": df_cfg_g2_without_sppf,
        "rsm": df_rsm_g2_without_sppf,
    },
).reset_index(level=0).rename(columns={"level_0": "Grammar"})
```

```python
import os

if not os.path.exists("pics/g2_without_sppf"):
    os.makedirs("pics/g2_without_sppf", exist_ok=True)
```

```python
for graph in graphs:
    plt.clf()
    bp = sns.violinplot(
        data = dff[(dff.graph == graph)],
        x="time_s",
        y="graph",
        hue="Grammar",
        palette=["m", "g"],
    )
    bp.set(xlabel='time(s)', ylabel='')
    plt.savefig(f"pics/g2_without_sppf/{graph}.png")
```
